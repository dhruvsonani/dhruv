```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel(r'C:\Users\a\Desktop\APMS Nihal & Ashutosh\Malay & Dhruv\Instance\1652\Model\1780_TTF.xlsx', usecols=['Date', 'Ch 278', 'Ch 279', 'Ch 280', 'Ch 281', 'Ch 282', 'Ch 283', 'Ch 284', 'Ch 285', 'Ch 303', 'Ch 324', 'Ch 325', 'Ch 326', 'Ch 327', 'Ch 328', 'Ch 329', 'Ch 330', 'Ch 331', 'Ch 1195', 'Ch 1196', 'Ch 1197', 'Ch 1198', 'Ch 1199', 'Ch 1200', 'Ch 1201', 'Ch 1202', 'Ch 1203', 'Ch 1204', 'Ch 1343', 'Ch 1344', 'Ch 1345', 'Ch 1346', 'Ch 1347', 'Ch 1627', 'Ch 1635', 'Ch 1639', 'Ch 1643', 'Ch 1644', 'Ch 1645', 'Ch 1646', 'Ch 1647', 'Ch 1648', 'Ch 1649', 'Ch 1650', 'Ch 1651', 'Ch 1652', 'Ch 1653', 'Ch 1654', 'Ch 1655', 'Ch 1656', 'Ch 1657', 'Ch 1658', 'Ch 1659', 'Ch 1660', 'Ch 1661', 'Ch 1663', 'Ch 1664', 'Ch 1665', 'Ch 1773', 'Ch 1778', 'Ch 1780', 'Ch 1817', 'Ch 1840', 'Instance', 'TTF'])
data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m-%d %H:%M:%S')

data
```

```python
# We'll drop columns which has only one unique value.

col_to_drop = []

for col in data.columns:
    if len(data[col].value_counts()) == 1:
        col_to_drop.append(col)

print(col_to_drop)

data_drop = data.drop(col_to_drop, axis=1)
```

```python
# Replace 'On'=>1 and 'Off'=>0

cat_cols = data_drop.select_dtypes(include=['object']).columns

for cat in cat_cols:
    data_drop[cat] = data_drop[cat].map({'Off':0, 'On':1})

data_drop
```

```python
# Now, we'll find instancewise missing values percentage.

null_values = data_drop.groupby('Instance').apply(lambda df: str(round(100*df.isnull().sum().sum()/(len(df)*39), 2))+'%').to_frame(name='NULL VALUES')

null_values
```

**Conclusion:**

So, there are 2 instances(218 & 222) which have null values. As instance 218 has very few missing values we can fill it by forward fill. And for 222 we'll get help of other channels to fill it's values.

```python
# Ch 218

na_values_218 = ['Ch 1343', 'Ch 1344', 'Ch 1345']

for col in na_values_218:
    data_drop[col] = data_drop[col].ffill()

print('Done!!!')
```

```python
num_cols = data.select_dtypes(include=['float64','int64']).columns[:-2]
```

```python
missing_values = data_drop.isna().sum()

miss_cols = missing_values[missing_values>0].index
```

```python
col_to_check = list(set(num_cols) - set(miss_cols))
```

```python
corr_check = []

for col in col_to_check:
    
    values = data_drop[col].value_counts()
    
    print('Column             :', col)
    print('Unique values      :', len(values))
    print()
    
    if len(values)>=12:
        corr_check.append(col)
```

```python
for col in corr_check:
    data_drop[col].value_counts().plot(kind='barh', figsize=(15,4))
    plt.title(col, fontweight='bold')
    plt.show()
```

We'll use **Ch 1196(Aft DG1 Phase1 Current)** to fill the missing values of other channels.

```python
import numpy as np

from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestRegressor
```

```python
data_missing_values = data_drop[list(miss_cols)+['Ch 1196']]

test_data = data_missing_values[data_missing_values.isnull().any(axis=1)]
train_data = data_missing_values[~data_missing_values.index.isin(test_data.index)]

train_data = train_data.sample(frac=1.0)
```

```python
X_missing_value = train_data[['Ch 1196']]
y_missing_value = train_data.drop(['Ch 1196'], axis=1)

X_missing_train, X_missing_valid, y_missing_train, y_missing_valid = train_test_split(X_missing_value, y_missing_value, test_size=0.15, random_state=21)
```

```python
model = RandomForestRegressor(random_state=21)
model.fit(X_missing_train, y_missing_train)

pred_train = model.predict(X_missing_train)
pred_valid = model.predict(X_missing_valid)

print('RMSE of Train :', np.sqrt(mean_squared_error(y_missing_train, pred_train)))
print('RMSE of Valid :', np.sqrt(mean_squared_error(y_missing_valid, pred_valid)))
```

```python
data_drop.loc[test_data.index,test_data.columns[:-1]]= model.predict(test_data[['Ch 1196']])

data_drop
```
