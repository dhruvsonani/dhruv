```python
import pickle
import pandas as pd
from itertools import combinations
```

```python
df = pd.read_csv(r'C:\Users\a\Desktop\APMS Nihal & Ashutosh\Malay & Dhruv\DG_1.csv')
df.head()
```

```python
col_with_single_value = []

for col in df.columns:
    if df[col].nunique() == 1:
        col_with_single_value.append(col)

print('>> Single values columns:', len(col_with_single_value))
print(col_with_single_value)

pickle.dump(col_with_single_value, open(r'C:\Users\a\Desktop\APMS Nihal & Ashutosh\Malay & Dhruv\Cleaned_Data\single_valued_columns.pkl', 'wb'))
```

```python
df['Ch 1279'].nunique()
```

```python
df = df.drop(col_with_single_value, axis=1)
```

```python
columns = df.columns[2:]
col_pairs = list(combinations(columns, 2))
```

```python
equal_cols = {}

for col1, col2 in col_pairs:
    if df[col1].equals(df[col2]):
        if col1 not in equal_cols:
            equal_cols[col1] = [col2]
        else:
            equal_cols[col1].append(col2)

print(equal_cols)
```

```python
pickle.dump(equal_cols, open(r'C:\Users\a\Desktop\APMS Nihal & Ashutosh\Malay & Dhruv\Cleaned_Data\equal_columns.pkl', 'wb'))
```

```python
cols_equal = []

for val in equal_cols.values():
    cols_equal.extend(val)

cols_equal = list(set(cols_equal))

print('>> Dropped Columns:', len(cols_equal))
print(cols_equal)
```

```python
df = df.drop(cols_equal, axis=1)
```

```python
total_cols = cols_equal + col_with_single_value
print('>> Total dropped values:', len(total_cols))
print(sorted(total_cols))
```
